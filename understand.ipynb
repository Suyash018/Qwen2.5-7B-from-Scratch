```python
from safetensors.torch import load_file
import torch

# Paths to the safetensor files
part1_path = "Qwen2.5-7B\model-00001-of-00004.safetensors"
part2_path = "Qwen2.5-7B\model-00002-of-00004.safetensors"
part3_path = "Qwen2.5-7B\model-00003-of-00004.safetensors"
part4_path = "Qwen2.5-7B\model-00004-of-00004.safetensors"

# Load each part
state_dict1 = load_file(part1_path)
state_dict2 = load_file(part2_path)
state_dict3 = load_file(part3_path)
state_dict4 = load_file(part4_path)

# Combine the state dictionaries
full_state_dict = {**state_dict1, **state_dict2, **state_dict3, **state_dict4}

# Now save the combined state dict as a new model
torch.save(full_state_dict, "Qwen2.5-7B\combined_model.pth")
```
